```python
import pandas as pd
import numpy as np

np.random.seed(42)
n = 1000

ages = np.random.randint(18, 71, n)
incomes = np.random.uniform(20000, 100000, n).round(2)
scores = np.random.rand(n)
genders = np.random.choice(['Male', 'Female'], n)
occupations = np.random.choice(['Engineer', 'Teacher', 'Doctor', 'Lawyer', 'Artist'], n)
cities = np.random.choice(['Jakarta', 'Bandung', 'Surabaya', 'Medan', 'Yogyakarta'], n)
has_car = np.random.choice([True, False], n)
num_children = np.random.randint(0, 6, n)
years_exp = np.random.randint(0, 41, n)

target = np.where((ages > 30) & (incomes > 50000), 'Yes', 'No')

df = pd.DataFrame({
    'age': ages,
    'income': incomes,
    'score': scores,
    'gender': genders,
    'occupation': occupations,
    'city': cities,
    'has_car': has_car,
    'num_children': num_children,
    'years_experience': years_exp,
    'target': target
})

print("=== 5 Baris Pertama dari Dataset ===")
display(df.head())

print("\n=== Informasi Dataset ===")
df.info()

print("\n=== Statistik Deskriptif (Kolom Numerik) ===")
display(df.describe())

for col in ['gender', 'occupation', 'city', 'has_car', 'target']:
    print(f"\n-- Value Counts untuk '{col}' --")
    display(df[col].value_counts())

print("\n=== Matriks Korelasi (Fitur Numerik) ===")
display(df.corr(numeric_only=True))
```
